# Step4:Data Cleaning/Feature Engineering

```python
alexa = alexa.drop(['date','rating'],axis = 1)
alexa
```

```python
variation_dummies = pd.get_dummies(alexa['variation'],drop_first = True)
variation_dummies
```

```python
#column variation will be replaced by variation_dummies 
#For the replacement of 'variation' with 'Dummies variation ' need to drop the 'variation' col
alexa.drop(['variation'],axis = 1,inplace = True)
alexa
```

```python
alexa = pd.concat([alexa,variation_dummies],axis = 1)
alexa
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
alexa_countvectorizer = vectorizer.fit_transform(alexa['verified_reviews'])
```

```python
alexa_countvectorizer.shape
#after vectorizing we get 4044 col for each wordpr
```

```python
print(vectorizer.get_feature_names())
```

```python
print(alexa_countvectorizer.toarray())
```

```python
#droping the column 'verified reviews to replace with'encoded reviews'.
alexa.drop(['verified_reviews'],axis = 1,inplace = True)
alexa
```

```python
encoded_reviews =pd.DataFrame(alexa_countvectorizer.toarray())
alexa = pd.concat([alexa,encoded_reviews],axis = 1)
alexa
```

```python
X = alexa.drop(['feedback'],axis = 1)
X.shape
```

```python
y = alexa['feedback']
y
```

# Step5:Model Training

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 5)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.ensemble import RandomForestClassifier
randomforest_classifier = RandomForestClassifier(n_estimators = 100,criterion = 'entropy')
randomforest_classifier.fit(X_train,y_train)
```

# Step6: Evaluating The Model

```python
y_prd_train = randomforest_classifier.predict(X_train)
cm = confusion_matrix(y_train,y_prd_train)
sns.heatmap(cm,annot = True, fmt = 'd')
```
