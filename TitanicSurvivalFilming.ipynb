# Step 4: Prepare the data for training/Data Cleaning

```python
sns.heatmap(trdt.isnull(),yticklabels = False,cbar = False,cmap = 'Blues')
```

```python
trdt.drop('Cabin',axis = 1,inplace = True)
```

```python
trdt.drop(['PassengerId','Name','Ticket','Embarked'],axis = 1,inplace = True)
```

```python
trdt
```

```python
sns.heatmap(trdt.isnull(),yticklabels = False,cbar = False,cmap = 'Blues')
```

```python
plt.figure(figsize=(15,10))
sns.boxplot(x = 'Sex',y = 'Age',data = trdt)
```

```python
def Fill_Age(data):
    age = data[0]
    sex = data[1]
    
    if pd.isnull(age):
        if sex is 'male':
            return 29
        else:
            return 25
        
    else:
        return age
```

```python
trdt['Age']= trdt[['Age','Sex']].apply(Fill_Age,axis = 1)
```

```python
sns.heatmap(trdt.isnull(),yticklabels = False,cbar = False,cmap = 'Blues')
```

```python
trdt['Age'].hist(bins = 20)
```

```python
male = pd.get_dummies(trdt['Sex'])
```

```python
male
```

```python
#to avoid dummy trape
male = pd.get_dummies(trdt['Sex'],drop_first = True)
male
```

```python
trdt.drop(['Sex'],axis = 1,inplace = True)
trdt
```

```python
trdt = pd.concat([trdt,male],axis = 1)
trdt
```

```python
X = trdt.drop('Survived',axis = 1).values
X
```

```python
y = trdt['Survived'].values
```

```python
y
```

# Model Training

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 0)
```

```python
#Fitting Logistic Regression to the training set
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train,y_train)
```

# Model Testing

```python
y_predict = classifier.predict(X_test)
y_predict
```

```python
from sklearn.metrics import classification_report,confusion_matrix
cm = confusion_matrix(y_test,y_predict)
sns.heatmap(cm,annot = True, fmt = 'd')
```

```python
print(classification_report(y_test,y_predict))
```
