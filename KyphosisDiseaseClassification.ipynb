# Training and Testing Dataset /Data Cleaning

```python
X = kpdt.drop(['Kyphosis'],axis = 1)
X
```

```python
y = kpdt['Kyphosis']
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train,y_train)
```

```python
feature_improtance = pd.DataFrame(decision_tree.feature_importances_,index = X_train.columns,columns = ['importance'])
feature_improtance
```

```python
feature_improtance = pd.DataFrame(decision_tree.feature_importances_,index = X_train.columns,columns = ['importance']).sort_values('importance',ascending = False)
feature_improtance
```

# Evaluating The Model

```python
from sklearn.metrics import classification_report,confusion_matrix
y_prd_train = decision_tree.predict(X_train)
y_prd_train
```

```python
cm = confusion_matrix(y_train,y_prd_train)
sns.heatmap(cm,annot = True, fmt = 'd')
print(classification_report(y_train,y_prd_train))
```

```python
y_prd_test = decision_tree.predict(X_test)
y_prd_test
```

```python
y_test
```

```python
cm = confusion_matrix(y_test,y_prd_test)
sns.heatmap(cm,annot = True, fmt = 'd')
```

```python
print(classification_report(y_test,y_prd_test))
```

# Improving The Model

```python
from sklearn.ensemble import RandomForestClassifier
randomforest_classifier = RandomForestClassifier(n_estimators = 100)
randomforest_classifier.fit(X_train,y_train)
```

```python
y_prd_train = randomforest_classifier.predict(X_train)
cm = confusion_matrix(y_train,y_prd_train)
sns.heatmap(cm,annot = True, fmt = 'd')
```

```python
y_prd_test = randomforest_classifier.predict(X_test)
y_prd_test
```

```python
cm = confusion_matrix(y_test,y_prd_test)
sns.heatmap(cm,annot = True, fmt = 'd')
```

```python
print(classification_report(y_test,y_prd_test))
```
