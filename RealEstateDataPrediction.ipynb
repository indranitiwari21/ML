```python
#Importing the independent data set 
X = df.iloc[:,:-1].values
X
```

```python
#importing the dependent data set
Y = df.iloc[:,6].values
Y
```

```python
#spliting the dataset into Training Set and Testing set
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 1/3,random_state = 0)
```

```python
# Model Building
```

```python
#Fitting Multiple Linear Regression to The Training Set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,Y_train)
```

```python
#Predicting The test set result
y_pred = regressor.predict(X_test)
y_pred
```

```python
"""We know that the equation of a mutliple linear regression line is given by y=b1+b2*x+b3*xâ€™+b4*xâ€+â€¦â€¦.
where b1, b2, b3, â€¦ are the coefficients and x, xâ€™, xâ€ are all independent variables.
Since we dont have any â€˜xâ€™ for the first coefficient we assume it can be written as a product of b and 1 and hence we append a column of ones. There are libraries that take care of it but since we are using the stats model library we need to explicitly add the column.""""
```

```python
# building the optimal model using backward elimination
# SL = 0.05 and eliminating those features which have p > SL
import statsmodels.formula.api as sm
X= np.append(arr = np.ones((414,1)).astype(int), values = X, axis = 1)
#np.append(x, x[:,-1:], axis=1)
```

```python
#independent variables that are statistically significant for dependent variable so now creating optimal matrix
X_opt = X[:,[0,1,2,3,4,5]]
regressor_ols = sm.OLS(endog = Y, exog = X_opt).fit()
```

```python
#Here important to know lower is the p-value the more significant our independent is going to be with respect to dependent variable. 
regressor_ols.summary()
```
