# Step4:Model Training

```python
bank = bank.drop(['Customer ID'],axis = 1)
bank
```

```python
X = bank.drop(['Retire'],axis = 1)
X
```

```python
y = bank['Retire']
y
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,y,test_size = 0.2,random_state = 5)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
Y_train.shape
```

```python
Y_test.shape
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report,confusion_matrix
```

```python
svc_model = SVC()
```

```python
svc_model.fit(X_train,Y_train)
```

# Step5:Evaluating The model

```python
y_prd = svc_model.predict(X_test)
y_prd
```

```python
cm = confusion_matrix(Y_test,y_prd)
sns.heatmap(cm,annot = True, fmt = 'd')
```

```python
print(classification_report(Y_test,y_prd))
#Result is not better in precision % so need to improve our model
```

# Step6:Improving The Model

```python
#Normalizing/linearization the datasets
min_train = X_train.min()
min_train
```

```python
range_train = (X_train - min_train).max()
range_train
```

```python
X_train_scaled = (X_train - min_train)/range_train
X_train_scaled
```

```python
#sns.scatterplot(x = X_train_scaled['age'],y = X_train_scaled['401k Savings'],hue = y_train)
```

```python
min_test = X_test.min()
```

```python
range_test = (X_test - min_test).max()
```

```python
X_test_scaled = (X_test - min_test)/range_test
X_test_scaled
```

```python
svc_model = SVC()
svc_model.fit(X_train_scaled,Y_train)
```

```python
y_prd = svc_model.predict(X_test_scaled)
cm = confusion_matrix(Y_test,y_prd)
sns.heatmap(cm,annot = True, fmt = 'd')
```

```python
print(classification_report(Y_test,y_prd))
```

# Part2:Improving The Model

```python
param_grid = {'C':[0.1,1,10,100],'gamma':[1,0.1,0.01,0.001],"kernel":['rbf']}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(SVC(),param_grid,refit = True,verbose = 4)
```

```python
grid.fit(X_train_scaled,Y_train)
```

```python
grid.best_params_
```

```python
grid_predict = grid.predict(X_test_scaled)
```

```python
cm = confusion_matrix(Y_test,grid_predict)
sns.heatmap(cm,annot = True, fmt = 'd')
```

```python
print(classification_report(Y_test,grid_predict))
```
